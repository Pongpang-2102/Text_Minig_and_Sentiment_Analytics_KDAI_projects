```python
import pandas as pd
import re

from pythainlp.tokenize import word_tokenize
import numpy as np

from gensim.models import Word2Vec
```

```python
file_name = 'train.csv'
file_name
```

```python
df = pd.read_csv(file_name,encoding ='utf-8',  usecols=[0,1] ,names=['sentence','sentiment'], header=0)
df.head()
```

```python
df.shape
```

```python
df.replace(r'[^‡∏Å-‡πô0-9]','',regex=True, inplace = True)
df.head()
```

```python
sentences = df.sentence.values
sentences.shape
```

```python
sentences[300:305]
```

```python
tsentences = []
for st in sentences:
    tokens = word_tokenize(st.strip())
    tsentences.append(tokens)

tsentences[900:905]
```

```python
dimension = 128
```

```python
w2v_model = Word2Vec(tsentences, min_count=1, vector_size=dimension, workers=6, sg=1, epochs=50)
```

```python
print(w2v_model)
```

```python
w2v_model.save('w2v_model.bin')
```

```python
new_model = Word2Vec.load('w2v_model.bin')
```

```python
new_model.wv.similarity('‡∏Å‡∏£‡∏∞‡∏ó‡∏∞', '‡∏´‡∏°‡∏π')
```

```python
new_model.wv.most_similar(positive=['‡πÄ‡∏ö‡∏µ‡∏¢‡∏£‡πå'], topn=5)
```

```python
#words = list(new_model.wv.vocab)
words = list(new_model.wv.index_to_key)
```

```python
words
```

```python
vectors = list(new_model.wv.vectors)
```

```python
vectors
```

```python
import csv

with open('w2vG.tsv', 'w') as out_file:
    tsv_writer = csv.writer(out_file, delimiter='\t')
    for v in vectors:
        tsv_writer.writerow(v)

with open('labelG.tsv', 'w') as out_file:
    tsv_writer = csv.writer(out_file, delimiter='\t')
    for word in words:
        tsv_writer.writerow([word])
```
