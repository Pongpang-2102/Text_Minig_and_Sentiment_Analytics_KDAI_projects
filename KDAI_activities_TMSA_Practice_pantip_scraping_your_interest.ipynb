```
#นำเข้าไลบรารี
from bs4 import BeautifulSoup
import pandas as pd
import requests
import re
```

```
#กำหนด URL ที่ต้องการดึงข้อมูล
url = 'https://pantip.com/search?q=eighty%20six'
response = requests.get(url)
```

```
#ใช้ BeautifulSoup ดึง HTML content
page_html = BeautifulSoup(response.content, 'html.parser')
```

```
print(page_html.prettify())
```

```
#ดึงข้อมูลที่ต้องการ
#movie_containers = page_html.find_all('ul', class_='pt-list pt-list-item__full-title pt-list__type-a')
movie_containers = page_html.find_all('li', class_='pt-list-item pt-list-item__sr__no-img')
movie_container_tag = page_html.find_all('li', class_='pt-list-item__tag')

links = []
subjects = []
members = [] # หมายเลข member เช่น  6739396
#like_score = []
# emotion_score = []
#date_post = []
tags = []
# comment = []
```

```
print(movie_containers)
```

```
print(movie_container_tag)
```

```
# For every movie of these 50
for container in movie_containers:
    # Scrape the name
    link = container.h2.a["href"]
    links.append(link)
    subject = container.h2.a.get_text(strip=True)
    subjects.append(subject)
    member = container.h5.a.get_text(strip=True)
    members.append(member)
    # tag = container.h2.get_text(strip=True)
```
