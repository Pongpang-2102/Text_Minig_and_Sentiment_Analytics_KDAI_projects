```python
df = pd.read_csv(SOURCE_File, usecols=[0,3,4] ,names=['id', 'topic','sentence'], header=0)
```

```python
df.head(10)
```

```python
qdf = df.copy()

#copy data frame
```

```python
# ตัดคำ
```

```python
qdf['clean_documents'] = qdf['sentence'].apply(lambda x: custom_tokenizer.word_tokenize(x))
```

```python
qdf.head()
```

```python
# เปลี่ยนรูป คำ ที่เก็บเป็น List เป็น Text โดยการต่อคำแต่ละคำด้วยช่องว่าง เพื่อให้ง่ายในการใช้งานครั้งต่อไป

# ใช้ map เพื่อเปลี่ยนค่าใน List บางค่าที่อาจไม่ใช่ String ให้เป็น String ก่อนการ join
```

```python
detokenized_doc = []
for i in range(len(qdf)):
    t = ' '.join(map(str, qdf['clean_documents'][i]))
    detokenized_doc.append(t)
qdf['clean_documents'] = detokenized_doc
```

```python
qdf.head()
```

```python
# ตัดคำโดยใช้ช่องว่าง
```

```python
# tokenization
tokenized_doc = qdf['clean_documents'].fillna('').apply(lambda x: x.split())
tokenized_doc
```

```python
# ตรวจสอบ Stop word ที่ Load มา
```

```python
stopwords = thai_stopwords()
len(stopwords)
```

```python
stopwords
```

```python
with open(THAI_STOP_WORDS) as f:
    new_stopwords_list = [line.strip() for line in f]

new_stopwords_list
```

```python
#stopwords = stopwords.union(new_stopwords_list)
# ไม่ใช้ Stop word ที่ Laod มา ใช้ของเราเท่านั้น

stopwords = new_stopwords_list
len(stopwords)
stopwords
```

```python
tokenized_doc = tokenized_doc.apply(lambda x: [item for item in x if item not in stopwords])
```

```python
#tokenized_doc = tokenized_doc.apply(lambda x: [item for item in x if len(item) > 2])
```

```python
# อ่าน คำผิด และคำที่ถูกต้อง จากไฟล์
```

```python
with open(THAI_CORRECT_WORDS) as f:
    thai_correct_word_list = [line.strip() for line in f]

thai_correct_word_list
```

```python
# ดึงคู่คำ คำผิด และคำที่ถูกต้อง ลง Dict
```

```python
correct_pair={}

for pair in thai_correct_word_list:
    correct_pair[pair.split()[0].strip()]=pair.split()[1].strip()
```

```python
# แก้คำผิด
```

```python
doc = []
for tokenized_list in tokenized_doc:
    for index, word in enumerate(tokenized_list):
        if word in correct_pair:
            tokenized_list[index]=correct_pair[word]
    doc.append(tokenized_list)

tokenized_doc = pd.Series(doc)
```

```python
# de-tokenization
detokenized_doc = []
for i in range(len(qdf)):
    t = ' '.join(tokenized_doc[i])
    detokenized_doc.append(t)
qdf['clean_documents'] = detokenized_doc
```
